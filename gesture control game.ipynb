```python
#Counting Fingers only tutorials using open cv

#binding trackbar with video

import cv2
import numpy as np 
import math
from controls import handling , motion

#Read Camera
cap = cv2.VideoCapture(0)

while True:
    _,frame = cap.read()
    
    frame = cv2.flip(frame,2)
    frame = cv2.resize(frame,(600,500))
    
    # Get hand data from the rectangle sub window
    cv2.rectangle(frame, (0,1), (400,500), (0, 0, 0), 0)
    crop_image = frame[1:500, 0:400]
    
    hsv = cv2.cvtColor(crop_image, cv2.COLOR_BGR2HSV)
    hsv = cv2.GaussianBlur(hsv, (9,9), 0)
    
    lower_bound = np.array([25,70,80])
    upper_bound = np.array([49,255,255])
   
    #Creating Mask
    mask = cv2.inRange(hsv, lower_bound, upper_bound)
    #filter mask with image
    filtr = cv2.bitwise_and(crop_image, crop_image, mask=mask)
    
    
    #mask1  = cv2.bitwise_not(mask)
    #m_g = cv2.getTrackbarPos("Thresh", "Color Adjustments") #getting track bar value
    ret,thresh = cv2.threshold(mask,10,255,cv2.THRESH_BINARY)
    
    
    #findcontour(img,contour_retrival_mode,method)
    cnts,hier = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
    
    print("len===",len(cnts))
    # compute the center of the contour
    
    #an image moment is a certain particular weighted average (moment) of the image pixels
    for c in cnts:
        M = cv2.moments(c)
        try:
           x = int(M["m10"] // M["m00"])
           y= int(M["m01"] // M["m00"])
           
           cv2.circle(crop_image, (x, y), 7, (0, 0, 0), -1)
           #cv2.drawContours(crop_image, [c], -1, (50, 50, 150), 2)
           x1,y1,w1,h1=cv2.boundingRect(cnts[0])
           x2,y2,w2,h2=cv2.boundingRect(cnts[1])
           #cv2.circle(crop_image,(x1,y1),1,(255,0,125),-1)
           #cv2.circle(crop_image,(x2,y2),1,(255,0,125),-1)
           cv2.line(crop_image,(x1,y1),(x2,y2),255,2)
           #find distance between line
           distance=math.sqrt(((x2-x1)**2) + ((y2-y1)**2))
           distance=round((distance/300)*100,2)
           print("distance==",distance)
           slope = math.tan(((y2-y1)/(x2-x1)))*100#convert the slope into %
           slope = round(slope,2)
           print("slope==",slope)
           cv2.putText(crop_image,"Turning:"+str(slope)+"deg",(50,50),cv2.FONT_HERSHEY_SIMPLEX,0.5,(255,255,255),2)
           cv2.putText(crop_image,"Acceleration:"+(str(distance)),(50,150),cv2.FONT_HERSHEY_SIMPLEX,0.5,(0,255,0),2)
           
           
           if(distance>100):#limit distance to 100
                    distance=100
           if slope>100:#limit angle to 100
                    slope=100
           elif slope< -100:
                    slope=-100
           handling(distance)#set xAxis on Joystick to the slope %
           motion(slope)#set yAxis on Joystick to speed %
        
        except:
           pass
       
       
       
       
       
         
    cv2.imshow("Thresh", thresh)
    
    #cv2.imshow("filter==",filtr)
    cv2.imshow("Result", frame)

    key = cv2.waitKey(25) &0xFF
    if key == 27:
        break
cap.release()
cv2.destroyAllWindows()
           
print("Done")                
                
              
```
